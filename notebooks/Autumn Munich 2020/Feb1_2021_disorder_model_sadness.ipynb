### Here is the figure I originally sent you:
Black points and lines are the disorder model, purple-orange is the FK model. I extracted the squared magnetisations (using the one for the largest system size) from the FK model at each point in parameter space then converted that to defect density p with $p = 1 - \sqrt{m^2}$ and produced a disorder model at that defect density for comparison.

Unfortunately the picture is even worse as I had made an error in my calculation of the disorder density.

Originally my method of generating CDW states with uncorrelated disorder was:
    1) generate a CDW state
    2) flip each state with probability p

This is effectively interpolating between the m=1 and m=-1 CDW states so
$m = p*(-1) + (1-p)*(1)$

So I swapped out the method for:
1) generate a CDW state (m=1) and a completely random state (m=0)
2) for each spin, choose the one from random state with probability p and from the CDW state otherwise. 

With this method you instead get $m = p*0 + (1-p)*(1)$ so the difference is only a factor of two but after fixing that, the curves now entirely overlap!
